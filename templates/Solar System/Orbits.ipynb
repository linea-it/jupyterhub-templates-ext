# Órbitas – Sistema Solar

Template para visualizar ou simular órbitas (ex.: com matplotlib).

```python
import matplotlib.pyplot as plt
import numpy as np

# Exemplo: órbitas circulares aproximadas
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect('equal')
ax.plot(0, 0, 'yo', markersize=15, label='Sol')
for a in [0.4, 0.7, 1.0, 1.5]:
    theta = np.linspace(0, 2*np.pi, 100)
    ax.plot(a*np.cos(theta), a*np.sin(theta), 'b-', alpha=0.5)
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.legend()
ax.set_title('Órbitas (esquemático)')
plt.show()
```
